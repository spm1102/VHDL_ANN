```python
import torch


BATCH_SIZE = 32
EPOCHS = 30

device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
from load_data import *


train_dataset, test_dataset = take_dataset(download=True)

plot_data(train_dataset)
print(f"Train images: {len(train_dataset)}")
print(f"Test images : {len(test_dataset)}")

train_dataloader, test_dataloader = take_dataloader(
    train_data=train_dataset,
    test_data=test_dataset,
    batch_size=BATCH_SIZE
)
```

```python
from model import *


model1 = model1()
model2 = model2()
model3 = model3()

models = (model1, model2, model3)
```

```python
from helpers import train_loop, save_model_params


for model in models:
    train_loop(
        model=model,
        epochs=EPOCHS,
        train_dataloader=train_dataloader,
        test_dataloader=test_dataloader,
        device=device
    )
    best_path = f"{model.__class__.__name__}_best.pth"
    model.load_state_dict(torch.load(best_path))
    save_model_params(model)
```
